```python
#2022/4/4
import cv2
cap = cv2.VideoCapture(0,cv2.CAP_DSHOW) #0表示默认camera
cap.open(0)

while cap.isOpened():
    flag, frame = cap.read()
    
    cv2.imshow('my_window',frame)
    
    key_pressed = cv2.waitKey(60)
    print('按键为',key_pressed)
    if key_pressed == 27: #ESC
        break
cap.release()
cv2.destroyAllWindows()
```
